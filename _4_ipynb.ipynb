Напишите цикл `while` , который позволяет пользователю ввести число. К числу необходимо прибавлять 10 и результат – присваивать переменной с именем product. Цикл должен повторяться до тех пор, пока product меньше 100.

````python
product = 0

while product < 100:
    try:
        number = int(input("Введите число: "))
        product = number + 10
        print(f"Текущее значение product: {product}")
    except ValueError:
        print("Ошибка: введите целое число")

print("Цикл завершен")
````

Напишите цикл `while`, который просит пользователя ввести два числа. Числа должны быть сложены, и показана сумма. Цикл должен запрашивать у пользователя, желает ли он выполнить операцию еще раз. Если да, то цикл должен повториться, в противном случае он должен завершиться.

````python
repeat = 'д'
while repeat.lower() == 'д':
    num1 = float(input("Введите первое число: "))
    num2 = float(input("Введите второе число: "))
    total = num1 + num2
    print(f"Сумма чисел: {total}")
    repeat = input("Хотите продолжить? (д/н): ")
print("Программа завершена.")
````

Напишите цикл, который просит пользователя ввести число. Цикл должен выполнить 10 итераций и вести учет нарастающего итога введенных чисел.

````python
total = 0
for i in range(1, 11):
    while True:
        try:
            number = float(input(f"Введите число {i} из 10: "))
            break
        except ValueError:
            print("Ошибка! Введите число.")
    total += number
    print(f"Текущий итог: {total}")

print(f"\nОбщая сумма: {total}")
````

**Сожженные калории.** Бег на беговой дорожке позволяет сжигать 4,2 калорий в минуту. Напишите программу, которая применяет цикл для вывода количества калорий, сожжен­ ных после 10, 15, 20, 25 и 30 минут бега.

````python

calories_per_minute = 4.2
minutes = [10, 15, 20, 25, 30]
print("Минуты\tКалории")
print("----------------")
for time in minutes:
    calories_burned = calories_per_minute * time
    print(f"{time}\t{calories_burned:.1f}")
````

**Уровень океана.** Допустим, что уровень океана в настоящее время повышается пример­но на 1,6 мм в год. С учетом этого создайте приложение, которое выводит количество миллиметров, на которые океан будет подниматься каждый год в течение следующих 25 лет.


````python

rise_per_year = 1.6
total_years = 25
print("Год\tПовышение уровня (мм)")
print("--------------------------")
for year in range(1, total_years + 1):
    total_rise = rise_per_year * year
    print(f"{year}\t{total_rise:.1f}")
````

Например, если поезд движется со скоростью 90 км/ч в течение трех часов, то пройден­ное расстояние составит 270 км. Напишите программу, которая запрашивает у пользова­теля скорость транспортного средства (в км/ч) и количество часов, которое оно двига­лось. Затем она должна применить цикл для вывода расстояния, пройденного транспорт­ным средством для каждого часа этого периода времени.
